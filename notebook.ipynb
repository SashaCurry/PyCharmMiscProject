**Задача 3**:
- Создать массив чисел от $-4\Pi$ до $4\Pi$ включительно, количество точек 100;
- Посчитать сумму поэлементных квадратов синуса и косинуса для данного массива;
- С помощью *np.all* проверить, что все элементы равны единице.

```python
pis = np.linspace(-4 * np.pi, 4 * np.pi, 100)
# print(pis)
trig_identity = (np.sin(pis)**2 + np.cos(pis)**2).round()
print(trig_identity)
print(np.all(trig_identity == 1.))
```

**Задача 4**: Создать квадратную матриуц размером 8, на главной диагонали арифметическая прогрессия с шагом 3 (начиная с 3), а на побочной -1. Остальные элементы 0.

```python
diag_matrix = np.diag(np.arange(3, 3*8 + 1, 3))
rev_diag_matrix = [-1 * lst[::-1] for lst in np.eye(8, dtype="int")]
main_matrix = diag_matrix + rev_diag_matrix
print(main_matrix)
```

```python
print(-1 * np.eye(8, dtype="int")[::-1] + np.diag(range(3, 3*8 + 1, 3)))
```
