**Задача 3**:
- Создать массив чисел от $-4\Pi$ до $4\Pi$ включительно, количество точек 100;
- Посчитать сумму поэлементных квадратов синуса и косинуса для данного массива;
- С помощью *np.all* проверить, что все элементы равны единице.

```python
pis = np.linspace(-4 * np.pi, 4 * np.pi, 100)
# print(pis)
trig_identity = (np.sin(pis)**2 + np.cos(pis)**2).round()
print(trig_identity)
print(np.all(trig_identity == 1.))
```

**Задача 5**: Отобразить вектор размера 100, в котором вычёркивается *x*, если *x* --- составное (т.е. не является простым)

```python
array = np.zeros(100, dtype=bool)
array[0] = True
simples = np.empty(0, dtype=int)
for i in range (2, 101):
    is_simple = True
    for simple in simples:
        if i % simple == 0:
            is_simple = False
            break
    if is_simple:
        array[i - 1] = True
        simples = np.append(simples, i)

nums = range(1, 101)
print(list((nums[i], bool(array[i])) for i in range(100)))
```

## Линейная алгебра

```python
a = np.array([[2, 1], [2, 3]])
print(a)
```

**Определитель матрицы**

```python
print(np.linalg.det(a))
```

**Обратная матрица**

```python
b = np.linalg.inv(a)
print(b)
```

```python
print(a.dot(b))
print(b.dot(a))
```

```python
c = np.array([[2, 1], [6, 3]])
print(c)
print(np.linalg.det(c))
```

```python
np.linalg.inv(c)
```

**Решение неоднородных линейных уравнений (НЛУ)**
$$A * x = v$$

```python
v = np.array([5, -10])
print(np.linalg.solve(a, v))
print(np.linalg.inv(a).dot(v))
```

**Собственные вектора матрицы $A$**
$$A * x  = \lambda * x$$

```python
l, u = np.linalg.eig(a)
print(l)
print(u)
```

**Собственные значений матрицы $A$ и $A.T$ сопадают**

```python
l, u = np.linalg.eig(a.T)
print(l)
print(u)
```
